```python
import pandas as pd
```

```python
FREQ = 480
DIR = "runs/1/"
dt = pd.read_csv(DIR + "csvData.csv", header=None)
```

```python
dt = dt.iloc[:, 1:]
dt = dt.transpose()
dt
```

```python
ch_names = ["Fp1", "Fp2", "F7", "F3", "Fz", "F4", "F8", "T3", "C3", "Cz", "C4", "T4", "P3", "P4", "O1", "O2"]
df2 = pd.DataFrame()

for c in range(len(ch_names)):
    df2[ch_names[c]] = dt[c]
df2 = df2 -512
df2 = df2 / 1024
```

```python
df2
```

```python
import mne
```

```python
#ch_names = ["Fp1-F7", "Fp1-F3", "F7-T3", "T3-O1", "F3-C3", "C3-P3", "P3-O1", "Fz-CZ", "Fp2-F4", "Fp2-F8", "F4-C4", "C4-P4", "P4-O2", "F8-T4", "T4-O2", "Number"]
ch_names = ["Fp1", "Fp2", "F7", "F3", "Fz", "F4", "F8", "T3", "C3", "Cz", "C4", "T4", "P3", "P4", "O1", "O2"]
info = mne.create_info(ch_names=ch_names, sfreq=FREQ, ch_types="eeg")
raw = mne.io.RawArray(df2.transpose(), info)

montage = 'standard_1020' # Electrode position file
raw.set_montage(montage)
```

```python
raw.compute_psd(fmax=50).plot()
```

```python
low_cut = 4
hi_cut  = 80

raw_filt = raw.copy().filter(low_cut, hi_cut)

raw_filt.compute_psd(fmax=80).plot()

```

```python
raw_filt.plot(start=20, duration=2, scalings="auto", show_scrollbars=False)
```

```python
# Filter settings
ica_low_cut = 0.1       # For ICA, we filter out more low-frequency power
hi_cut  = 80

raw_ica = raw.copy().filter(ica_low_cut, hi_cut)
```

```python
tstep = 1.0
events_ica = mne.make_fixed_length_events(raw_ica, duration=tstep)
epochs_ica = mne.Epochs(raw_ica, events_ica,
                        tmin=0.0, tmax=tstep,
                        baseline=None,
                        preload=True)
```

```python
from autoreject import get_rejection_threshold

reject = get_rejection_threshold(epochs_ica)
reject
```

```python
# ICA parameters
random_state = 42   # ensures ICA is reproducable each time it's run
ica_n_components = .99     # Specify n_components as a decimal to set % explained variance

# Fit ICA
ica = mne.preprocessing.ICA(n_components=ica_n_components,
                            random_state=random_state,
                            )
ica.fit(epochs_ica,
        reject=reject,
        tstep=tstep)
```

```python
ica.plot_components()
```

```python
raw_ica = ica.apply(raw_ica)
```

```python
raw_ica.plot(start=41, duration=2, scalings="auto")
```
